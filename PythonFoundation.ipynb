```python
# Inheritance: Multiple Inheritance, Multi Level Inheritance
class A: 
    def method(self):
        print("from A")
        

class B(A):
    def method(self):
        print("from B")
        
        
class C(B):
    def method(self):
        print("from C")
        
        
class D(B, A):
    def method(self):
        print("from D")
        super().method()
```

### Statistical Analysis

```python
# Chi Square Test
```


    Single category: one-proportion test:
    Two categories: Chi Squared test: basically frequency comparison between categories wrt a certain column like male vs female over height, male vs female over education
    One numeric: t test
    One numeric and one categorical: t-test (30 samples least) or ANOVA if 3 categories/groups to compare

```python
# Q1. Is there any difference between men and women salary ? General Question: More than or less Than does nto matter
# Two Tail test
```

```python
data_interim = df[["sex", "earning"]]
```

```python
data_cont = pd.crosstab(data_interim["sex"], data_interim["earning"])
```

```python
data_cont # contingency table
```

```python
from scipy.stats import chi2_contingency
```

```python
stat, p, dof, expected = chi2_contingency(data_cont)
```

```python
alpha = 0.05

if p <= alpha:
    
    print('Dependent (reject H0)')
    
else:
    print('Independent (H0 holds true)')
```

```python
# chi-squared test with similar proportions
from scipy.stats import chi2_contingency
from scipy.stats import chi2
# contingency table
#table = [ [], [13984,  6396]]
table = [[10,20,30], [6,8,17]]
print(table)
stat, p, dof, expected = chi2_contingency(table)
print('dof=%d' % dof)
print(expected)
# interpret test-statistic
prob = 0.95
critical = chi2.ppf(prob, dof)
print('probability=%.3f, critical=%.3f, stat=%.3f' % (prob, critical, stat))
if abs(stat) >= critical:
    print('Dependent (reject H0)')
else:
    print('Independent (fail to reject H0)')
# interpret p-value
alpha = 1.0 - prob
print('significance=%.3f, p=%.3f' % (alpha, p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (fail to reject H0)')
```
